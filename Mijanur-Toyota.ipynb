```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x='Age', y='Price',color="violet", data=df)
plt.title('Scatter Plot of Age vs Price')
plt.xlabel('Age')
plt.ylabel('Price')
plt.show()
```

```python
plt.hist(df["KM"], bins=10, color='violet', edgecolor='Red')
plt.title('Histogram of KM')
plt.xlabel('KM')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
sns.kdeplot(df["Age"], shade=True, color='Violet')
plt.title('Density Plot of Age')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
sns.boxplot(df["Price"],color="Violet")
plt.title('boxplot of Price')
plt.xlabel("Frequency")
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
sns.countplot(df["FuelType"],hue="Automatic",data=df)
plt.title('Countplot of FuelType')
plt.xlabel("FuelType")
plt.ylabel('Automatic')
plt.grid(True)
plt.show()
```
